```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models
import shap
import numpy as np
import matplotlib.pyplot as plt


def build_ncf_model(input_dim):
    inputs = layers.Input(shape=(input_dim,))
    x = layers.Dense(128, activation='relu')(inputs)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    model = models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

ncf_model = build_ncf_model(X_train.shape[1])
history = ncf_model.fit(X_train, y_train,
                        validation_data=(X_test, y_test),
                        epochs=20,
                        batch_size=256,
                        verbose=2)

loss, accuracy = ncf_model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

print("Generating SHAP explanations...")
explainer = shap.Explainer(ncf_model, X_train)
shap_values = explainer(X_test[:100])


print("Displaying SHAP Summary Plot...")
shap.summary_plot(shap_values, X_test[:100], feature_names=X.columns)

print("Displaying SHAP Bar Plot...")
shap.summary_plot(shap_values, X_test[:100], feature_names=X.columns, plot_type="bar")

print("\nRetention Strategies:")
print("- For customers with high churn risk due to low usage, offer incentives like discounts for higher usage plans.")
print("- For those with long subscription lengths but high churn risk, send loyalty benefits or surveys to understand their concerns.")
print("- Analyze specific high-impact features (like location or age group) and tailor retention campaigns accordingly.")



def build_seq2seq_model(input_shape):
    encoder_inputs = layers.Input(shape=(input_shape[1], input_shape[2]))
    encoder = layers.LSTM(64, return_state=True)
    encoder_outputs, state_h, state_c = encoder(encoder_inputs)
    encoder_states = [state_h, state_c]

    decoder_inputs = layers.Input(shape=(None, input_shape[2]))
    decoder_lstm = layers.LSTM(64, return_sequences=True, return_state=True)
    decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
    decoder_dense = layers.Dense(1, activation='sigmoid')
    decoder_outputs = decoder_dense(decoder_outputs)

    model = models.Model([encoder_inputs, decoder_inputs], decoder_outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```
